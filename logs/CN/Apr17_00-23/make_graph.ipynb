```python
import matplotlib.pyplot as plt
import numpy as np

# Path to your log file
filepath = 'logger'
file = open(filepath, 'r')
lines = file.readlines()
# List to store mean rewards
mean_rewards = []

# Read the file
count = 1
for line in lines:
    # Find the start of the mean reward section
    start = line.find('mean reward')
    if start != -1:
        start += 12
        # Find the end which is the start of std reward section
        end = line.find(',', start)
        # Extract and convert the mean reward to float
        mean_reward = float(line[start:end].strip())
        # Append to the list
        if len(mean_rewards) == 0 or mean_reward > mean_rewards[-1] - 10:
            mean_rewards.append(mean_reward)
            count += 1

# Plotting the mean rewards
plt.figure(figsize=(10, 5))
plt.plot(np.arange(1, count), mean_rewards, color='b')
plt.title('Mean Reward Over Time')
plt.xlabel('Training Step')
plt.ylabel('Mean Reward')
plt.grid(True)


# Path to your log file
filepath = 'logger_base'
file = open(filepath, 'r')
lines = file.readlines()
# List to store mean rewards
mean_rewards_base = []

# Read the file
count = 1
for line in lines:
    # Find the start of the mean reward section
    start = line.find('mean reward')
    if start != -1:
        start += 12
        # Find the end which is the start of std reward section
        end = line.find(',', start)
        # Extract and convert the mean reward to float
        mean_reward = float(line[start:end].strip())
        # Append to the list
        if len(mean_rewards_base) == 0 or mean_reward > mean_rewards_base[-1] - 10:
            mean_rewards_base.append(mean_reward)
            count += 1
plt.plot(np.arange(1, count), mean_rewards_base, color='b')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to read mean rewards from a logfile
def read_mean_rewards(filepath):
    file = open(filepath, 'r')
    lines = file.readlines()
    mean_rewards = []
    count = 0  # To track the number of entries added for plotting

    for line in lines:
        start = line.find('mean reward')
        if start != -1:
            start += 12  # Adjust to start reading after 'mean reward'
            end = line.find(',', start)  # End of the mean reward value
            mean_reward = float(line[start:end].strip())  # Convert to float
            # Append to list if it's the first value or meets the condition
            if len(mean_rewards) == 0 or mean_reward > mean_rewards[-1] - 10:
                mean_rewards.append(mean_reward)
                count += 1

    return np.arange(1, count+1), mean_rewards  # Return X and Y for plotting

# Read mean rewards from both files
x1, y1 = read_mean_rewards('logger')
x2, y2 = read_mean_rewards('logger_base')

# Plotting both mean rewards
plt.figure(figsize=(10, 5))
plt.plot(x1, y1, label='2WayComm', color='b')  # Label for the first line
plt.plot(x2, y2, label='Baseline', color='r')  # Label for the second line

# Add a title, axis labels, and grid
plt.title('Mean Reward Over Time')
plt.xlabel('Training Step')
plt.ylabel('Mean Reward')
plt.grid(True)

# Add a legend to the plot to differentiate the lines
plt.legend()

# Show the plot
plt.show()
```
